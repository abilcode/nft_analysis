```python
content = tweet_data.copy()
content['post'] = content['username'] 
content = content[['username','post','nlikes','nretweets']].groupby(['username'],as_index=False).agg({'post':'count','nlikes':'sum','nretweets':'sum'})
content['power'] = (content.nlikes + 1.5*content.nretweets)/content.post
```

```python
g=sns.barplot(x="username", y="power", data=content.sort_values('power',ascending=0).head(10))
plt.title('Content Power Analysis')
plt.xticks(rotation=90)
content.sort_values('power',ascending=0,inplace=True)
```

```python
user = [u for u in content.username.head(20)]
for u in user:
    print(tweet_data[tweet_data['username']==u]['tweet'])
```

```python
import emoji as emj
import regex

EMOJIS = emj.UNICODE_EMOJI["en"]
tweet = pd.DataFrame(tweet_data['tweet'])
tweet
def extract_emoji(df):
    df["emoji"] = ""
    for index, row in df.iterrows():
        for emoji in EMOJIS:
            if emoji in row["tweet"]:
                row["tweet"] = row["tweet"].replace(emoji, "")
                row["emoji"] += emoji
extract_emoji(tweet)
tweet

```
